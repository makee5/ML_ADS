# Create an imbalanced dataset

### First, we will create an imbalanced data set from the iris data set.

```python
from sklearn.datasets import load_iris

from imblearn.datasets import make_imbalance

iris = load_iris(as_frame=True)

sampling_strategy = {0: 10, 1: 20, 2: 47}
X, y = make_imbalance(iris.data, iris.target, sampling_strategy=sampling_strategy)
```

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(ncols=2, figsize=(10, 5))
autopct = "%.2f"
iris.target.value_counts().plot.pie(autopct=autopct, ax=axs[0])
axs[0].set_title("Original")
y.value_counts().plot.pie(autopct=autopct, ax=axs[1])
axs[1].set_title("Imbalanced")
fig.tight_layout()
```

## Performing over-sampling using SMOTE

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE()
X_resampled, y_resampled = smote.fit_resample(X, y)

# pie chart of resampled data
fig, axs = plt.subplots(ncols=2, figsize=(10, 5))
y_resampled.value_counts().plot.pie(autopct=autopct, ax=axs[0])
axs[0].set_title("Resampled")
y.value_counts().plot.pie(autopct=autopct, ax=axs[1])
axs[1].set_title("Imbalanced")
fig.tight_layout()
```

## Performing under-sampling using ClusterCentroids

```python
from imblearn.under_sampling import ClusterCentroids

tomek = ClusterCentroids()
X_resampled, y_resampled = tomek.fit_resample(X, y)

# pie chart of resampled data
fig, axs = plt.subplots(ncols=2, figsize=(10, 5))
y_resampled.value_counts().plot.pie(autopct=autopct, ax=axs[0])
axs[0].set_title("Resampled")
y.value_counts().plot.pie(autopct=autopct, ax=axs[1])
axs[1].set_title("Imbalanced")
fig.tight_layout()
```
