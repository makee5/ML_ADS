### Task Types

Classification tasks are tasks where the target variable is categorical. There are two types of classification tasks: binary classification and multi-class classification. In binary classification tasks, the target variable has two categories. In multi-class classification tasks, the target variable has more than two categories.

In the iris dataset, the target variable has three categories. Therefore, it is a multi-class classification task. In the breast cancer dataset, the target variable has two categories. Therefore, it is a binary classification task.

Regression tasks are tasks where the target variable is continuous. In the diabetes dataset, the target variable is continuous. Therefore, it is a regression task.

```python
# Classification vs Regression

# load classification dataset
from sklearn.datasets import load_iris

iris = load_iris()
iris_X = iris.data
iris_y = iris.target
iris_y
```

```python
# pie plot of the target variable
import matplotlib.pyplot as plt

plt.pie([len(iris_y[iris_y==0]), len(iris_y[iris_y==1]), len(iris_y[iris_y==2])], labels=iris.target_names, autopct='%1.1f%%')
plt.show()
```

```python
from sklearn.datasets import load_diabetes

diabetes = load_diabetes()
diabetes_X = diabetes.data
diabetes_y = diabetes.target
diabetes_y
```

```python
# histogram plot of the target variable
plt.hist(diabetes_y, bins=60)
plt.show()
```

```python
# Binary Classification vs Multi-Class Classification

# load binary classification dataset
from sklearn.datasets import load_breast_cancer

cancer = load_breast_cancer()
cancer_X = cancer.data
cancer_y = cancer.target
cancer_y
```

```python
# pie plot of the target variable
plt.pie([len(cancer_y[cancer_y==0]), len(cancer_y[cancer_y==1])], labels=cancer.target_names, autopct='%1.1f%%')
plt.show()
```

### Classification Metrics

Classification metrics are used to evaluate the performance of classification models. There are several classification metrics. Some of the most commonly used classification metrics are accuracy, precision, recall, F1 score, and ROC AUC score.

They compare the predicted values with the actual values and calculate the performance of the model. 

```python
import numpy as np

true_values = cancer.target
dummy_predicitions = np.random.choice([0, 1], size=len(true_values))
dummy_predicitions
```

```python
# Confusion Matrix
from sklearn.metrics import confusion_matrix

confusion_matrix(true_values, dummy_predicitions)
```

```python
# Accuracy
accuracy = np.mean(dummy_predicitions == true_values)
accuracy
```

```python
# with sklearn
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(true_values, dummy_predicitions)
accuracy
```

```python
# Precision
precision = np.sum((dummy_predicitions == 1) & (true_values == 1)) / np.sum(dummy_predicitions == 1)
precision
```

```python
# with sklearn
from sklearn.metrics import precision_score

precision = precision_score(true_values, dummy_predicitions)
precision
```

```python
# Recall
recall = np.sum((dummy_predicitions == 1) & (true_values == 1)) / np.sum(true_values == 1)
recall
```

```python
# with sklearn
from sklearn.metrics import recall_score

recall = recall_score(true_values, dummy_predicitions)
recall
```

```python
# F1 Score
f1 = 2 * (precision * recall) / (precision + recall)
f1
```

```python
# with sklearn
from sklearn.metrics import f1_score

f1 = f1_score(true_values, dummy_predicitions)
f1
```

```python
# classification report
from sklearn.metrics import classification_report

print(classification_report(true_values, dummy_predicitions))
```

```python
# ROC AUC Score
from sklearn.metrics import roc_auc_score

roc_auc = roc_auc_score(true_values, dummy_predicitions)
roc_auc
```

```python
# plot ROC curve
from sklearn.metrics import roc_curve

fpr, tpr, thresholds = roc_curve(true_values, dummy_predicitions)

plt.plot(fpr, tpr)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.show()
```

### Multi-Class Classification Metrics

In multi-class classification tasks, the same metrics can be used. However, they can be calculated for each class separately. Generally, the average of these metrics is used to evaluate the performance of the model. However, the average can be calculated in different ways. The most commonly used average types are micro, macro, and weighted.

- Micro: Aggregating statistics across all classes, treating each instance equally.

- Macro: Calculating metrics per class and averaging, treating all classes equally.

- Weighted: Taking a class-specific average, with weights based on class frequency.

```python
# confusion matrix
true_values = iris.target
dummy_predicitions = np.random.choice([0, 1, 2], size=len(true_values))

confusion_matrix(true_values, dummy_predicitions)
```

```python
# f1 score with micro, macro, and weighted average
f1_micro = f1_score(true_values, dummy_predicitions, average='micro')
f1_macro = f1_score(true_values, dummy_predicitions, average='macro')
f1_weighted = f1_score(true_values, dummy_predicitions, average='weighted')
f1_micro, f1_macro, f1_weighted
```

```python
# classification report
print(classification_report(true_values, dummy_predicitions))
```

### Regression Metrics

Regression metrics are used to evaluate the performance of regression models. There are several regression metrics. Some of the most commonly used regression metrics are mean absolute error (MAE), mean squared error (MSE), root mean squared error (RMSE), and R-squared score.

```python
true_values = diabetes.target
dummy_predicitions = np.random.choice(diabetes.target, size=len(true_values))
dummy_predicitions
```

```python
# Mean Absolute Error
mae = np.mean(np.abs(dummy_predicitions - true_values))
mae
```

```python
# with sklearn
from sklearn.metrics import mean_absolute_error

mae = mean_absolute_error(true_values, dummy_predicitions)
mae
```

```python
# Mean Squared Error
mse = np.mean((dummy_predicitions - true_values) ** 2)
mse
```

```python
# with sklearn
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(true_values, dummy_predicitions)
mse
```

```python
# Root Mean Squared Error

rmse = np.sqrt(mse)
rmse
```

```python
# with sklearn
from sklearn.metrics import root_mean_squared_error

rmse = root_mean_squared_error(true_values, dummy_predicitions)
rmse
```

```python
# R-squared Score
ss_res = np.sum((true_values - dummy_predicitions) ** 2)
ss_tot = np.sum((true_values - np.mean(true_values)) ** 2)
r_squared = 1 - (ss_res / ss_tot)
r_squared
```

```python
# with sklearn
from sklearn.metrics import r2_score

r_squared = r2_score(true_values, dummy_predicitions)
r_squared
```

## Error Estimation Methods

Error estimation methods are used to estimate the performance of a model. The most commonly used error estimation methods are holdout method and cross-validation method.

## Holdout Method

The holdout method is the simplest error estimation method. It splits the dataset into two parts: training set and test set. The model is trained on the training set and evaluated on the test set. The performance of the model is estimated using the test set. The holdout method is simple and easy to implement. However, it may not provide an accurate estimate of the performance of the model, especially when the dataset is small.

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(iris_X, iris_y, test_size=0.2, random_state=12)
```

```python
# train and evaluate a model
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
accuracy
```

## Cross-Validation Method

The cross-validation method is a more reliable error estimation method compared to the holdout method. It splits the dataset into k parts, called folds. The model is trained on k-1 folds and evaluated on the remaining fold. This process is repeated k times, each time using a different fold as the test set. The performance of the model is estimated using the average of the performance on the test sets. The cross-validation method provides a more accurate estimate of the performance of the model, especially when the dataset is small.

```python
from sklearn.model_selection import cross_val_score

model = LogisticRegression()
accuracy = cross_val_score(model, iris_X, iris_y, cv=5, scoring='accuracy')
accuracy
```

```python
accuracy.mean()
```

```python
# cross-validation with different metrics
precision = cross_val_score(model, iris_X, iris_y, cv=5, scoring='precision_macro')
recall = cross_val_score(model, iris_X, iris_y, cv=5, scoring='recall_macro')
f1 = cross_val_score(model, iris_X, iris_y, cv=5, scoring='f1_macro')
precision.mean(), recall.mean(), f1.mean()
```
